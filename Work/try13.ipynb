```python
def find_All_href_Tag() :
    for link in soup.findAll('a'):
        print(link.get('href'))
find_All_href_Tag()
```

```python

def find_all_classname():
    for li in soup.find_all(class_="plp-carousel__stage plp-carousel__stage--1"):
        print(li.a.get('href'))
find_all_classname()
```

```python

def find_All_data_product():
    for div in soup.findAll('span'):
        print (div.get('data-product'))
        
find_All_data_product()    
```

```python

def find_all_wrapper():
    for li in soup.find_all(class_="plp-item"):
        print(li.a.get('href'))
find_all_wrapper()
```
