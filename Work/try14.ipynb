```python

def find_all_wrapper():
    
    a = []
    for li in soup.find_all(class_="plp-item"):
        a.append(li.a.get('href'))
    return a


df = find_all_wrapper()


print('total Result: ' )
df

# output_df = pd.DataFrame(find_all_wrapper())
# output_df.to_csv('some_name.csv')
```
