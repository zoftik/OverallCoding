```python
import pandas as pd
import win32com.client as wd
conn = wd.Dispatch(r'ADODB.Connection')
conn.CommandTimeout = 3600
CONNECTIONSTRING = "Provider=MSOLAP.5;Integrated Security=SSPI;Persist Security Info=True;Initial Catalog=ODS_2;Data Source=ssas.kotakseconline.com;MDX Compatibility=1;Safety Options=2;MDX Missing Member Mode=Error"
conn.Open(CONNECTIONSTRING)
rs = wd.Dispatch(r'ADODB.Recordset')

mdx_query6 = """
Select {[Measures].[Brokerage],[Measures].[Strike Plus Premium Turnover]}*
{[Segment-Hierarchy].&[Cash],[Segment-Hierarchy].&[Derivative]}
on Columns,


Non Empty{[Client Code].children*[Month Name].children}
on Rows 

From [ods logs] 

where 
({[Date].[FiscalPeriod].[Fiscal Year].[FY 2023].[2023 FQ 2]})

"""

dataset6 = rs.Open(mdx_query6, conn, 0, 1)
t2 = rs.GetRows()
df6 = pd.DataFrame(list(t2))
t_df6 = df6.transpose()
t_df6.columns = ['client_cd','Month','Brok_cash','Brok_Der','Vol_cash','Vol_Der']

t_df6.fillna(0, inplace=True)
t_df6.head()
t_df6.shape
```

```python
df_Brok_cash= pd.pivot_table(t_df6, index=['client_cd'], columns=('Month'),values=("Brok_cash"),aggfunc=np.sum)
df_Brok_cash.reset_index(inplace=True)
df_Brok_cash.fillna(0, inplace=True)
df_Brok_cash.head()
df_Brok_cash.shape
```
